# H1 Free and Reduced Lunch Rates in Nebraska from 2013-2018

This is a poor visual representation of the growth rate of free and reduced lunch within the state of Nebraska.

```r
ggplot(nefree, aes(x=DISTRICTLNAME)) + geom_bar()
```

Although you cannot individually read district names on this initial chart it is at least visually evident that there is a number of schools with a high amount of children on free and reduced lunch in the state of Nebraska. Almost all of the schools with a count under 10 children on free and reduced lunch are within more rural areas or smaller communities were poverty and lack of finacial assistance is next to nothing.

```r
districts <- nefree %>%
  group_by(DISTRICTLNAME) %>%
  summarise(
    count = n(),
  ) %>% arrange(desc(count))
```

```r
head(districts)
```

```r
ggplot(districts, aes(DISTRICTLNAME, weight=count)) + geom_bar()
```

```r
districts <- nefree %>%
  group_by(DISTRICTLNAME) %>%
  summarise(
    count = n(),
  ) %>% arrange(desc(count)) %>% 
 filter(count > 10)
```

```r
ggplot(districts, aes(DISTRICTLNAME, weight=count)) + geom_bar()
```

> This chart was meant to display the top 10 schools with the highest count of free and reduced lunch in the state of Nebraska. As you can tell it eveidently did not work.

As you can somewhat clearly read the schools with the highest free and redued lunch rates within Nebraska are elementry school. As the years progress and more children are born into poverty these numbers continue to rise with a grow amount of children elementry age accounting for the highest percentages of free and reduced lunch.

```r
ggplot(districts, aes(DISTRICTLNAME, weight=count)) + geom_bar()+ ggtitle("Nebraska Schools with the Highest Free & Reduced Lunch") + 
     theme(plot.title = element_text(lineheight=.8,  face="bold",)) + theme_light() +
labs(x="District", y="Free & Reduced Lunch Count", subtitle="Nebraksa Schools in more urban areas have a significantly larger portion of children on free and reduced lunch.", caption="Source: NE Data | By Maddie Johnson")+ coord_flip()
```

This poor representative chart fails to properly display the existential amount of children in Nebraska on free and reduced lunch.  When looking at the data properly the chart should display the large majority of urban schools in the state with over 50% of students on free or reduced lunch.
